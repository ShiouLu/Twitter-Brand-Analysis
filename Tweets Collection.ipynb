# <br>
# 2. Extracting data from Twitter
<br>

```python
#-- create some parameters for the tweepy search
search_terms = "deliveroo,Deliveroo" 
date_until = "2022-04-29"
no_items = 1000

#-- collect tweets
tweets = tweepy.Cursor(twitter_api.search_tweets,
              q=search_terms,
              lang="en",
              until=date_until).items(no_items)

#-- create a list to store them in
results = []

#-- loop around each tweet 
for tweet in tweets:

    #-- print the current tweet's timestamp and first 75 characters
    print(tweet.created_at, tweet.text[:75])
    
    #-- add the tweet to our results
    results.append(tweet)
```

# <br>
# 3. Storing the Tweets into a CSV file
<br>

```python
#-- Open/create a file to append data to 
csvFile = open('deliveroo_test.csv', 'a+', encoding='utf-8')

#-- use a CSV writer object to save the file 
csvWriter = csv.writer(csvFile)

#-- loop around each tweet we downloaded...
for tweet in results:
    
    #-- Write a row to the CSV file using some specific fields
    csvWriter.writerow([tweet.author.id,
                        tweet.author.name,
                        tweet.author.screen_name,
                        tweet.author.verified,
                        tweet.author.description,
                        tweet.author.location,
                        tweet.author.followers_count, 
                        tweet.author.friends_count, 
                        tweet.author.listed_count,
                        tweet.author.favourites_count, 
                        tweet.author.statuses_count,
                        tweet.author.lang,
                        tweet.created_at,
                        tweet.text,
                        tweet.retweet_count,
                        tweet.entities.get('hashtags'),
                        tweet.source,tweet.favorite_count])

print("Tweets appended to a file")
csvFile.close()
```
